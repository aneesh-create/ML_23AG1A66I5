```python
import pandas as pd
DataSet=pd.read_csv('Name-Age-Marks-City.csv')
DataSet.head()
```

```python
DataSet.fillna({
    'Age':DataSet['Age'].mean(),
    'Marks':DataSet['Marks'].mean(),
    'City':'Unknown'
},inplace=True)
print(DataSet)
```

```python
import pandas as pd
#Create example data including duplicates
data = {
    'Name':['Alice','Bob','Alice'],
    'Age':[18,18,18],
    'Gender':['F','M','F'],
    'Marks':[85, 90, 85]
}
df = pd.DataFrame(data)
print("original Data:")
print(df)
```

```python
#Remove duplicates across all columns
df_no_duplicates = df.drop_duplicates()
print("\nAfter Removing Duplicates:")
print(df_no_duplicates)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['John', 'Alice'],
    'Age': ['25', '30'],
    'Salary': ['50000', '55000']
})
print(df.dtypes)
```

```python
import pandas as pd
df = pd.DataFrame({
   'Name':['aneesh','krishna'],
   'Age':['25','30'],
   'Salary':['5000','8000']
})
print(df.dtypes)
```

convert specific columns to desired types

```python

df['Age']= df['Age'].astype(int)    #To integer
df['Salary'] = df['Salary'].astype(float)  #To float
print(df.dtypes)
```

if your data may have invalid values((numbers and words)),force conversion and set errors to "NaN":

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
print(df.dtypes)
```
